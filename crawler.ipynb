```python
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.PhantomJS('/Users/lunab/Documents/phantomjs-2.1.1-macosx/bin/phantomjs')
driver.implicitly_wait(3)
```

```python
# 최대 1000
start_page = 1000

min_num = 10000000000
table = []

last_max_num = 15911412
stop = False

for page in range(1,start_page+1):
    
    if stop:
        break
    
    driver.get('https://movie.naver.com/movie/point/af/list.nhn?&page='+str(page))
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    tbody = soup.find('tbody')
    
    list_num = tbody.select('td.ac')
    list_point = tbody.select('td.point')
    list_movie = tbody.select('a.movie')
    list_content = tbody.select('td.title')

    # debug 
    print('page',page)

    if page == 1:
        last_max_num = int(list_num[i].text)
    
    for i in range(10):
        num = int(list_num[i].text)
        
        if num <= last_min_num:
            stop=True

        # 크롤링도중 생기는 데이터로 인한 밀림으로 생기는 중복 방지
        if min_num <= num:
            print(page,i,num,'continue')
            continue

        min_num = num

        point = list_point[i].text
        movie = list_movie[i].text
        content = list_content[i].contents[4:5][0].strip()

        # 혹시 몰라 num 과 movie를 저장..
        table.append([num, movie, point, content])
```
